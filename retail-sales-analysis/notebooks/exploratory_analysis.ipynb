# Análisis Exploratorio de Ventas al por Menor

Este notebook se utiliza para realizar un análisis exploratorio interactivo del conjunto de datos de ventas al por menor. A continuación, se cargarán los datos y se realizarán diversas exploraciones.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
# Añade después de las otras importaciones
from src.exploration import calcular_total_ventas_por_categoria, calcular_promedio_ventas_diarias

# Añadir directorio raíz al path para encontrar los módulos
sys.path.append(os.path.abspath('..'))

# Opción 1: Cargar directamente con pandas (recomendado)
ruta = '../data/retail_sales_dataset.csv'
datos_df = pd.read_csv(ruta)

# Mostrar las primeras filas del DataFrame
datos_df.head()
```

## Análisis de Ventas por Categoría

A continuación, calcularemos el total de ventas por categoría.

```python
# Calcular el total de ventas por categoría
total_ventas_categoria = calcular_total_ventas_por_categoria(datos_df)

# Visualizar el total de ventas por categoría
plt.figure(figsize=(10, 6))
sns.barplot(x=total_ventas_categoria.index, y=total_ventas_categoria.values)
plt.title('Total de Ventas por Categoría')
plt.xlabel('Categoría de Producto')
plt.ylabel('Total de Ventas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calcular el total de ventas por categoría
total_ventas_categoria = calcular_total_ventas_por_categoria(datos_df)

# Visualizar el total de ventas por categoría como gráfico circular
plt.figure(figsize=(10, 8))
plt.pie(total_ventas_categoria.values, 
        labels=total_ventas_categoria.index, 
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        explode=[0.05] * len(total_ventas_categoria))  # Separar ligeramente todas las porciones

plt.title('Distribución de Ventas por Categoría', fontsize=14)
plt.axis('equal')  # Para que el gráfico sea circular
plt.tight_layout()
plt.show()
```

## Promedio de Ventas Diarias

Calcularemos el promedio de ventas diarias.

```python
# Calcular el promedio de ventas diarias
promedio_ventas_diarias = calcular_promedio_ventas_diarias(datos_df)
promedio_ventas_diarias

# Visualizar el promedio de ventas diarias
plt.figure(figsize=(10, 6))
sns.lineplot(data=promedio_ventas_diarias)
plt.title('Promedio de Ventas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Promedio de Ventas')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)  # Añadir cuadrícula para mejor legibilidad
plt.tight_layout()  # Ajustar el diseño para evitar recortes
plt.show()
```
